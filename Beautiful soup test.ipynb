```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = r'http://www.pythonforbeginners.com'
```

```python
r = urllib.request.urlopen(url).read()
```

```python
data = str(r)
```

```python
soup = BeautifulSoup(data, "lxml")
```

```python
print(soup.prettify())
```

```python
print(soup.title.string)
```

```python
print(soup.p)
```

```python
print(soup.a)
```

```python
for links in soup.find_all('a'):
    print(links)
```
